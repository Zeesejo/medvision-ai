# 01 — Data Exploration
Explore NIH Chest X-Ray14 dataset statistics, class distribution, and demographic breakdown.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

CSV_PATH = '../data/raw/nih_chestxray/Data_Entry_2017.csv'
df = pd.read_csv(CSV_PATH)
print(df.shape)
df.head()
```

```python
# Disease label distribution
from collections import Counter
all_labels = []
for labels in df['Finding Labels']:
    all_labels.extend(labels.split('|'))
counts = Counter(all_labels)
label_df = pd.DataFrame(counts.items(), columns=['Disease', 'Count']).sort_values('Count', ascending=False)
sns.barplot(data=label_df, x='Count', y='Disease')
plt.tight_layout()
plt.savefig('../results/figures/label_distribution.png', dpi=150)
plt.show()
```

```python
# Demographic breakdown
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
df['Patient Gender'].value_counts().plot.pie(ax=axes[0], autopct='%1.1f%%', title='Gender')
df['Patient Age'].plot.hist(ax=axes[1], bins=30, title='Age Distribution')
plt.tight_layout()
plt.savefig('../results/figures/demographics.png', dpi=150)
plt.show()
```
